```python
# dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set x-axis and variables
x_axis = np.arange(1,10,0.1)
print(x_axis)

sin = np.sin(x_axis)
cos = np.cos(x_axis)

print(sin)
print(cos)
```

```python
# draw a horizontal line with 0.25 transparency
# value for alpha can only go up to 1
plt.hlines(0,0,10, alpha = .5)
```

```python
# assign plots to tuples to be able to access it later; that stores the result of the plot
# ()

# calling this variable below will change the properties
sine_handle, = plt.plot(x_axis, sin, marker='x', color='green', label='Sine')

cos_handle, = plt.plot(x_axis, cos, marker='*', color='orange', label='Cosine')

# magic plot will continue to add new series to the same interactive plot, until it is saved as a jpeg
```
