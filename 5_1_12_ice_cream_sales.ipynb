```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
```

```python
# Tell matplotlib to create a scatter plot based upon the above data
# (x values, y values, s = sales, c = color, alpha = transparency value [0 to 1])
# y values should be a function (dependent) on x values
# e.g., ice cream sales (y-values) are dependent on the temperature (x-values)
plt.scatter(temp,sales,s=sales,c='pink',alpha=0.5)
```

```python
# Set the upper and lower limits of our y axis
plt.ylim(0,max(sales)+100)
```

```python
# Set the upper and lower limits of our x axis
plt.xlim(10,max(temp)+5)
```

```python
# Create a title, x label, and y label for our chart
plt.title('Ice Cream Sales v Temp')
plt.xlabel('Temp')
plt.ylabel('Sales')
plt.show()
```
