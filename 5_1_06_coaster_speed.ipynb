```python
# dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
danger = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]

time = np.arange(0,130,10)
time
```

```python
danger_handle, = plt.plot(time, danger, marker='x', color='red', label='Danger Drop')
rail_handle, = plt.plot(time, rail_gun, marker='s', color='green', label='RailGun')

plt.xlabel('Coaster Runtime')
plt.ylabel('Speed (MPH)')
plt.title('Coaster Speed Over Time')
plt.grid()
plt.legend(loc='best')
```
