```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
# Average weather temp
# created list with same number of values as x_axis_data
# if they have different numbers of values, you get an error like: 
#  x and y must have same first dimension, but have shapes (12,) and (11,)
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Plot the line
plt.plot(x_axis_data, points)
plt.show()
```

```python
# Convert to Celsius C = (F-32) * 0.56
# list comprehension, same as for loop
# for x in points:
#     (x-32) * 0.56
points_C = [(x-32) * 0.56 for x in points]
points_C
```

```python
# Plot both on the same chart
plt.plot(x_axis_data, points)
plt.plot(x_axis_data, points_C)
plt.show()
```

```python
# change colors and add markers and labels
far_handle, = plt.plot(x_axis_data,points,marker='x',color='red',label='Fahrenheit')
cel_handle, = plt.plot(x_axis_data,points_C,marker='s',color='blue',label='Celsius')
```

```python
# add labels to plt
plt.legend(loc='best')

# axes titles
plt.xlabel('Months')
plt.ylabel('Degrees')
```
