```python
# Dependencies
import matplotlib.pyplot as plt # graphing library
import numpy as np # statistical library (although other libraries can have functions for stats as well)
import pandas as pd # dataframe library
```

```python
# Load in csv
file = '../Resources/avg_rain_state.csv'
df = pd.read_csv(file)
df.head()
```

```python
# for x-values, store name of states to variable 'state' from dataframe as a series
state = df['State']

# just confirming the data type as a series; not needed to make chart
print(type(state)) #<class 'pandas.core.series.Series'>

# preview data
state.head()
```

```python
# for y-values, store rain in inches to variable 'rain' from dataframe as a series
rain = df['Inches']
rain.head()
```

```python
# stores "a range" of values from min=0 to max=len(state) by increments of 1
# the state series has 50 states in it, so the list of values goes up to 49
x_axis = np.arange(0,len(state),1)

# printing x_axis to see values generated by np.arange; [0,1,2,...49]
print(x_axis)

# plot bar chart with x_axis and bars for rainfall in each state; no labels so far
plt.bar(x_axis,rain)

# list comprehension to generate list of values, one per bar which will generate a tick each
tick_location = [value for value in x_axis]

# printing tick_location to see what is generated; not needed for chart
# [0,1,2,3...49]
print(tick_location)

# plot ticks by tick_location (list of sequential numbers up to but not including the number of values in x)
# add state labels and label rotation (in degrees)
plt.xticks(tick_location,labels=state,rotation=45)

# while not necessary for rendering chart in Jupyter Notebook (may be needed to visualize chart in other platforms)
# plt.show() does keep from printing extra numbers/text along with chart - keeps output clean
plt.show()
```

```python
# add title and axes labels
plt.title('Rain per State')
plt.xlabel('State')
plt.ylabel('Rain (inches)')
```
